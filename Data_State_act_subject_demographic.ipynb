```python
#packages needed
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True,
                  'axes.titlepad': 20})
```

```python
#creating dataframes from csv's
df_2022 = pd.read_csv('datasets/cae_2022.csv')
df_2021 = pd.read_csv('datasets/cae_2021.csv')
```

I am exploring in this notebook are:

How does the average ACT score for each subject over the years progressed in Kentucky State? By Demographic



```python
#Filtering  the data that meets criteria!.

df_state_2022 = df_2022[(df_2022['DISTRICT NAME'] == 'State')  ] 
df_state_2021 = df_2021[(df_2021['DISTRICT NAME'] == 'State')  ] 
```

```python
#Choosing  the columns that needed
df_state_subject_2022 = df_state_2022[['DEMOGRAPHIC', 'Average ACT Scores: English' , 'Average ACT Scores: Reading', 'Average ACT Scores: Math', 'Average ACT Scores: Science', 'Average ACT Composite Score' ]]
df_state_subject_2021 = df_state_2021[['DEMOGRAPHIC', 'Average ACT Scores: English' , 'Average ACT Scores: Reading', 'Average ACT Scores: Math', 'Average ACT Scores: Science', 'Average ACT Composite Score' ]]
```

```python
#Removing the null values from the dataframes
df_state_subject_2022  = df_state_subject_2022.dropna()
df_state_subject_2021  = df_state_subject_2021.dropna()
```

```python
df_state_subject_2022
```

```python
df_state_subject_2021
```

```python
#Merging both 2021 and 2022 school data 

df_state_subject_trend = pd.merge(df_state_subject_2021, df_state_subject_2022, how = 'inner' ,  on = ['DEMOGRAPHIC'] )

df_state_subject_trend.head()
```

```python
#Renaming the columns to respective years
df_state_subject_trend.rename(columns = {'Average ACT Scores: English_x':'Average ACT Scores: English_2021', 'Average ACT Scores: English_y' : 'Average ACT Scores: English_2022'}, inplace = True)
df_state_subject_trend.rename(columns = {'Average ACT Scores: Reading_x':'Average ACT Scores: Reading_2021', 'Average ACT Scores: Reading_y' : 'Average ACT Scores: Reading_2022'}, inplace = True)
df_state_subject_trend.rename(columns = {'Average ACT Scores: Math_x':'Average ACT Scores: Math_2021', 'Average ACT Scores: Math_y' : 'Average ACT Scores: Math_2022'}, inplace = True)
df_state_subject_trend.rename(columns = {'Average ACT Scores: Science_x':'Average ACT Scores: Science_2021', 'Average ACT Scores: Science_y' : 'Average ACT Scores: Science_2022'}, inplace = True)
df_state_subject_trend.rename(columns = {'Average ACT Scores: Science_x':'Average ACT Scores: Science_2021', 'Average ACT Scores: Science_y' : 'Average ACT Scores: Science_2022'}, inplace = True)
df_state_subject_trend.rename(columns = {'Average ACT Composite Score_x':'Average ACT Composite Score_2021', 'Average ACT Composite Score_y' : 'Average ACT Composite Score_2022'}, inplace = True)
```

```python
#changing the order for easy comparison
df_state_subject_trend = df_state_subject_trend.iloc[:, [0,1,6,2,7,3,8,4,9,5,10]]
df_state_subject_trend 
```

```python
#Pick the county name to find ACT Score for each subject by DEMOGRAPHIC
input_demographic  =  input("Enter the DEMOGRAPHIC Name")
print (input_demographic)
```

```python
#filtering the data by specified Demographic
df_state_demographic_subject_trend = df_state_subject_trend[(df_state_subject_trend['DEMOGRAPHIC'] == input_demographic) ] 
df_state_demographic_subject_trend 
```

The Average ACT Score Treding for each subject in Kentucky for a specificied demographic.

```python
#Plotting the Trending 
fig = df_state_demographic_subject_trend.plot(x= "DEMOGRAPHIC", kind="bar", figsize=(16,6) )
#fmt = '{x:,.0f}'
fig.yaxis.set_major_formatter('{x:,.0f}')
fig.set_title( f'The Average ACT Score for each subject for {input_demographic} students in Kentucky')
```
