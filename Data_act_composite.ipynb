```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True,
                  'axes.titlepad': 30})
```

```python
#creating dataframes from csv's
df_2022 = pd.read_csv('datasets/cae_2022.csv')
df_2021 = pd.read_csv('datasets/cae_2021.csv')
```

```python
#Print the sample data for 2022
df_2022.head()
```

```python
#Print the sample data for 2021
df_2021.head()
```

I am exploring in this notebook are:

 How does the average ACT Composite Score look over the years? By county and by
demographic

```python
#Filtering  the data that meets criteria!.
#df_2022 = df_2022[(df_2022['SCHOOL NAME'] == '---District Total---') & (df_2022['DISTRICT NAME'].str.contains('County')) & (df_2022['DEMOGRAPHIC']  == 'All Students') ] 
#df_2021 = df_2021[(df_2021['SCHOOL NAME'] == '---District Total---') & (df_2021['DISTRICT NAME'].str.contains('County')) & (df_2021['DEMOGRAPHIC']  == 'All Students') ] 

df_2022 = df_2022[(df_2022['SCHOOL NAME'] == '---District Total---') & (df_2022['DISTRICT NAME'].str.contains('County'))  ] 
df_2021 = df_2021[(df_2021['SCHOOL NAME'] == '---District Total---') & (df_2021['DISTRICT NAME'].str.contains('County'))  ] 
```

```python
#Choosing  the data that needed
df_2022 = df_2022[[ 'COUNTY NAME' , 'Average ACT Composite Score', 'DEMOGRAPHIC' ]]
df_2021 = df_2021[[ 'COUNTY NAME' , 'Average ACT Composite Score', 'DEMOGRAPHIC' ]]
```

```python
#sample data after filtering the data for 2022
df_2022.head()
```

```python
#sample data after filtering the data for 2021
df_2021.tail()
```

```python
#Removing the null values
df_2022 = df_2022.dropna()
df_2021 = df_2021.dropna()
```

```python
#sample data after cleaning the data for 2022
df_2022.head()
```

```python
#sample data after cleaning the data for 2021
df_2021.tail()
```

```python
#Merging both 2021 and 2022 school data 
Act_Composite_Trend = pd.merge(df_2021, df_2022, how = 'inner' ,  on = ['COUNTY NAME', 'DEMOGRAPHIC' ] )

Act_Composite_Trend.head()
```

```python
#Renaming the columns with respective year
Act_Composite_Trend.rename(columns = {'Average ACT Composite Score_x':'Average ACT Composite Score_2021', 'Average ACT Composite Score_y' : 'Average ACT Composite Score_2022' }, inplace = True)
Act_Composite_Trend.rename(columns = {'DEMOGRAPHIC_x':'DEMOGRAPHIC_2021', 'DEMOGRAPHIC_y' : 'DEMOGRAPHIC_2022' }, inplace = True)
```

```python
#sample data after renaming the columns
Act_Composite_Trend.head()
```

```python
fig = Act_Composite_Trend.plot(x= "COUNTY NAME", y=["Average ACT Composite Score_2021", "Average ACT Composite Score_2022"], kind="bar")
#fmt = '{x:,.0f}'
fig.yaxis.set_major_formatter('{x:,.0f}')
fig.set_title('ACT Composite Score by County')
```
